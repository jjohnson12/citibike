```python
# Dependencies
import pandas as pd
```

```python
# Import the data from CSV to Pandas

citibike2019_1 = pd.read_csv("Data/201901-citibike-tripdata.csv")
citibike2019_2 = pd.read_csv("Data/201902-citibike-tripdata.csv")
citibike2019_3 = pd.read_csv("Data/201903-citibike-tripdata.csv")
citibike2019_4 = pd.read_csv("Data/201904-citibike-tripdata.csv")
citibike2019_5 = pd.read_csv("Data/201905-citibike-tripdata.csv")
citibike2019_6 = pd.read_csv("Data/201906-citibike-tripdata.csv")
citibike2019_7 = pd.read_csv("Data/201907-citibike-tripdata.csv")
citibike2019_8 = pd.read_csv("Data/201908-citibike-tripdata.csv")
citibike2019_9 = pd.read_csv("Data/201909-citibike-tripdata.csv")
citibike2019_10 = pd.read_csv("Data/201910-citibike-tripdata.csv")
citibike2019_11 = pd.read_csv("Data/201911-citibike-tripdata.csv")
citibike2019_12 = pd.read_csv("Data/201912-citibike-tripdata.csv")
```

```python
# Join the dataframes

citibike2019 = citibike2019_1.append([citibike2019_2, citibike2019_3, citibike2019_4, citibike2019_5, citibike2019_6, citibike2019_7, 
                                      citibike2019_8, citibike2019_9, citibike2019_10, citibike2019_11, citibike2019_12])
```

```python
# Fix the column names

citibike2019.rename(columns={'tripduration': 'Trip Duration (seconds)', 'starttime': 'Start Time', 'stoptime': 'Stop Time', 
                             'start station id': 'Start Station ID' , 'start station name': 'Start Station Name', 
                             'start station latitude': 'Start Station Latitude', 'start station longitude': 'Start Station Longitude', 
                             'end station id': 'End Station ID', 'end station name': 'End Station Name', 'end station latitude': 'End Station Latitude', 
                             'end station longitude': 'End Station Longitude', 'bikeid': 'Bike ID', 'usertype': 'User Type', 
                             'birth year': 'Birthyear', 'gender': 'Gender'}, inplace = True)
```

```python
# Check random sample of the data

citibike2019.sample(5)
```

```python
# Add rider age

citibike2019['Age'] = 2020 - citibike2019['Birthyear']

citibike2019.sample(5)
```

```python
# Change gender from int to string
# Zero = unknown, one = male, two = female

citibike2019['Gender'].replace(0, 'Unknown',inplace=True)
citibike2019['Gender'].replace(1, 'Male',inplace=True)
citibike2019['Gender'].replace(2, 'Female',inplace=True)

citibike2019.sample(5)
```

```python
# Save data as Pickle

citibike2019.to_pickle("Data/citibike2019.pkl")
```

```python
# Read the Pickle

citibike2019 = pd.read_pickle("Data/citibike2019.pkl")
```

```python
# Check dataset size
citibike2019.info()
```

```python
# Remove rows with any missing data.

citibike2019.dropna(how = 'any', inplace = True)

# Check the new dataframe size.
citibike2019.info()
```

```python
# Creating a random sample of 14,999,999 since tableau public won't handle sets larger than this.

citibike2019set = citibike2019.sample(14999999)
```

```python
# Save to CSV

citibike2019set.to_csv("Data/citibike2019set.csv", index = True)
```

```python
# Save to JSON

citibike2019.to_json("Data/citibike2019.csv", index = False)
```
